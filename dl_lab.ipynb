````python

import numpy as np
def neuron(x, w, theta):
    if np.dot(x, w) >= theta:
        return 1
    else:
        return 0
#and
w_and = np.array([1, 1])
theta_and = 2
for x_input in [(0, 0), (0, 1), (1, 0), (1, 1)]:
    print(x_input, neuron(np.array(x_input), w_and, theta_and))
#or
w_or = np.array([1, 1])
theta_or = 1
for x_input in [(0, 0), (0, 1), (1, 0), (1, 1)]:
    print(x_input, neuron(np.array(x_input), w_or, theta_or))
print(f" Pankaj (421) 6aimlb1")
````

````python
import numpy as np

def art_neuron(x, w, b, activation="step"):
    z = np.dot(x, w) + b
    if activation == "step":
        if z >= 0:
            return 1
        else:
            return 0
    elif activation == "sigmoid":
        return 1 / (1 + np.exp(-z))
    elif activation == "relu":
        return max(0, z)
    else:
        return None

x = np.array([6, 2])
w = np.array([3, 5])
b = 3


print("step output", art_neuron(x, w, b, "step"))
print("sigmoid output", art_neuron(x, w, b, "sigmoid"))
print("relu output", art_neuron(x, w, b, "relu"))
print(f"Pankaj (421) 6 aimlb1")
````

````python
import numpy as np
# for AND gate (Input/Output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 7
w,b = np.zeros(2),0

for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    print("error : ",E,"<- at epoch",_)
    w += E*x
    b += E
    print(w,b)

````

````python
import numpy as np
# for AND gate (Input/Output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 5
w,b = np.random.rand(2), np.random.rand()

for _ in range(epochs):
  total_error_in_epoch = 0 # Initialize total error for the current epoch
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    total_error_in_epoch += abs(E) # Accumulate absolute error
    print("error : ",E,"<- at epoch",_)
    w += E*x
    b += E
    print(w,b)
  if total_error_in_epoch == 0:
    print(f"Converged at epoch {_}. No errors found.")
    break # Stop training if no errors occurred in the epoch
````
